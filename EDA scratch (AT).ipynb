```python
edit = edit['site_code year month day hour minute second time_decimal value value_std_dev nvalue latitude longitude altitude elevation intake_height qcflag'].str.split(' ', expand=True)
```

```python
column_names = ['site_code', 'year', 'month', 'day', 'hour', 'minute', 'second', 'time_decimal', 'value', 'value_std_dev', 'nvalue', 'latitude', 'longitude', 'altitude', 'elevation', 'intake_height', 'qcflag']
```

```python
edit.columns = column_names
```

```python
edit
```

```python
edit['day'].replace('1', '01', inplace=True)
edit['day'].replace('2', '02', inplace=True)
edit['day'].replace('3', '03', inplace=True)
edit['day'].replace('4', '04', inplace=True)
edit['day'].replace('5', '05', inplace=True)
edit['day'].replace('6', '06', inplace=True)
edit['day'].replace('7', '07', inplace=True)
edit['day'].replace('8', '08', inplace=True)
edit['day'].replace('9', '09', inplace=True)

edit['month'].replace('1', '01', inplace=True)
edit['month'].replace('2', '02', inplace=True)
edit['month'].replace('3', '03', inplace=True)
edit['month'].replace('4', '04', inplace=True)
edit['month'].replace('5', '05', inplace=True)
edit['month'].replace('6', '06', inplace=True)
edit['month'].replace('7', '07', inplace=True)
edit['month'].replace('8', '08', inplace=True)
edit['month'].replace('9', '09', inplace=True)
```

```python
edit['year-month-day'] = edit['year'] + '-' + edit['month'] + '-' + edit['day']
```

```python
edit.drop(columns=['year', 'month', 'day', 'hour', 'minute', 'second'], inplace=True)
```
